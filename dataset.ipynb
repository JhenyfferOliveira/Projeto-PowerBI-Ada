### Código usado para criar colunas de identificação + unir datasets

```python
import pandas as pd

df_prime = pd.read_csv('AmazonPrime\\titles_AmazonPrime.csv')
df_netflix = pd.read_csv('Netflix\\titles_Netflix.csv')
```

```python
df_prime.head()
```

```python
# Excluir colunas com valores nulos
excluir_colunas = ['age_certification', 'imdb_id']
df_prime = df_prime.drop(excluir_colunas, axis=1)
df_prime.columns
```

```python
df_netflix.head()
```

```python
# Criando função para criar coluna 'streaming' e preencher com 'Amazon Prime'
# Possui 9868 linhas
def inserir_coluna_prime(coluna):
    nova_coluna = 'Amazon Prime'
    coluna['streaming'] = nova_coluna

    return coluna
```

```python
df_prime = df_prime.apply(lambda x : inserir_coluna_prime(x), axis = 1)
df_prime.head()
```

```python
# Excluir colunas com valores nulos
excluir_colunas = ['age_certification', 'imdb_id']
df_netflix = df_netflix.drop(excluir_colunas, axis=1)
df_netflix.columns
```

```python
# Criando função para criar coluna 'streaming' e preencher com 'Netflix'
# Possui 5850 linhas
def inserir_coluna_netflix(coluna):
    nova_coluna = 'Netflix'
    coluna['streaming'] = nova_coluna

    return coluna
```

```python
df_netflix = df_netflix.apply(lambda x : inserir_coluna_netflix(x), axis = 1)
df_netflix.head()
```

```python
df_dataset_junto = pd.concat([df_prime, df_netflix], ignore_index=True)
df_dataset_junto
```

```python
df_dataset_junto.to_csv('amazon_netflix.csv')
```

```python
# Limpeza de dados
excluir_colunas = ['Unnamed: 0', 'id', 'description', 'runtime']
df_novo_dataset = df_novo_dataset.drop(excluir_colunas, axis=1)
df_novo_dataset.columns
```

```python
df_novo_dataset.fillna(-1, inplace=True)
```

```python
#dados = pd.read_csv(caminho_arquivo)
import uuid

def gerar_id():
    return str(uuid.uuid4())[:4]

df_novo_dataset['ID'] = [gerar_id() for _ in range(len(df_novo_dataset))]

#print(dados)
df_novo_dataset.head()
```
